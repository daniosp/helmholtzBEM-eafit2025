# Incident Field Computation (Solutions for Helmholtz Eq.)

#### Libraries

```python
%matplotlib widget
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

---

```python
# Screen width and height
upper_grid_size = 20
lower_grid_size = -upper_grid_size
ngrid_pts = 500

grid_vec = np.linspace(lower_grid_size,upper_grid_size,num=ngrid_pts)

X,Y = np.meshgrid(grid_vec,grid_vec)
```

```python
k = 4
```

```python
# Define the first field
k_ui_1 = k
ui_field_1 = np.sin(k_ui_1*X)

# Plot the surface
plt.close('all')

fig, ax = plt.subplots()
plot = ax.pcolormesh(X, Y, ui_field_1, edgecolor="none",cmap="copper",vmin=-1, vmax=1)
ax.axis('equal')
ax.set_aspect('equal', 'box')

cbar = fig.colorbar(plot,ax=ax,ticks=[-1,0,1])
plt.title(r"Incident field 1 ($u_i$)")
plt.show()
```

```python
# Define the second field
k_ui_2 = k
ui_field_2 = np.sin(k_ui_2*Y)

# Plot the surface
plt.close('all')

fig, ax = plt.subplots()
plot = ax.pcolormesh(X, Y, ui_field_2, edgecolor="none",cmap="copper",vmin=-1, vmax=1)
ax.axis('equal')
ax.set_aspect('equal', 'box')

cbar = fig.colorbar(plot,ax=ax,ticks=[-1,0,1])
plt.title(r"Incident field 2 ($u_i$)")
plt.show()
```

```python
# Define the third field
k_ui_3 = k
ui_field_3 = np.sin(k_ui_3*Y) + np.cos(k_ui_3*X)

# Plot the surface

fig, ax = plt.subplots()
plot = ax.pcolormesh(X, Y, ui_field_3, edgecolor="none",cmap="copper",vmin=-1, vmax=1)
ax.axis('equal')
ax.set_aspect('equal', 'box')

cbar = fig.colorbar(plot,ax=ax,ticks=[-1,0,1])
plt.title(r"Incident field 3 ($u_i$)")
plt.show()
```
