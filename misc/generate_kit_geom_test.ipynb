```python
import matplotlib.pyplot as plt
import numpy as np
```

Generate the geometry with a parametric eq. 



```python
alpha = 0.65
beta = 1.5

t = np.linspace(0,2*np.pi,200)

x = np.cos(t) + alpha * np.cos(2*t)
y = beta * np.sin(t)
```

```python
# Plot the surface
plt.close('all')

fig, ax = plt.subplots()
plot = ax.plot(x, y)
ax.axis('equal')

ax.set_aspect('equal', 'box')
plt.title(r"Incident field $u_i$")
plt.show()
```
